## 1. Why is Logistic Regression termed as Regression and not classification?

There are two main things that logistic regression termed as regression not classification.

The main difference between regression and classification is the type of target variable they deal with. In regression, the target variable is numerical or continuous, while in classification, the target variable is categorical or discrete.

Logistic Regression is a type of classification algorithm used for supervised learning. It is called Logistic Regression because it is based on the principles of regression analysis, which is a statistical method used for predicting continuous values.

## 3. What do you understand by Pruning in a Decision Tree and Why do we require Pruning in Decision Trees? Explain.

Decision tree is popular tool for classification and prediction. A decision teree is a flow chart like tree structure and it works for both categorical and continuous input and output variable.

Pruning is the process of removing branches from the tree that do not provide any additional information or lead to overfitting.
There are different pruning techniques, such as pre-pruning and post-pruning, which determine when and how to stop growing the tree and which nodes to remove.

We require Pruning in Decision tree for various purpose like:

To avoid overfitting : When building decision trees, they can become too detailed and complex, trying to fit every single data point, including irrelevant or random patterns. This is like memorizing the training data instead of learning from it. Pruning is a technique that prevents this overfitting problem by simplifying the tree. It removes unnecessary details and focuses on the most important patterns. This makes the tree easier to understand and more reliable for making predictions on new data

To Improve model interpretability : When a decision tree is pruned, it becomes simpler and easier to understand. Unnecessary parts of the tree that do not help in making accurate predictions are removed. This makes the tree less complicated and easier to interpret. With a pruned tree, we can see the important factors that influence the decisions and understand the logic behind them more clearly. This improved interpretability allows us to gain better insights from the model and explain its decision-making process more effectively to others.

Computational efficiency : Pruned trees are generally smaller and have fewer nodes and branches. This reduces the computational resources required for training, prediction, and storage. Pruning can help create more efficient decision tree models, especially when dealing with large datasets.


## 4. What do you understand about Information Gain? Also, explain the mathematical formulation associated with it.


Information gain is a measure used in machine learning to determine the usefulness of a feature to
predicting the target variable. It is calculated by comparing the entropy of the target variable before and
after splitting on a feature. The higher the information gains, the more effective the feature is in
predicting the target variable.
