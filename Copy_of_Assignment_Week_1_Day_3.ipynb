## 1. What is multicollinearity, and how do you treat it and when is it not a problem?

Multicollinearity is a situation in machine learning in which one independent variables(features) are highly correlated with other independent variables.

**Handling multicolinearity:**

Variable selection: Identify and remove one or more of the correlated variables from the model.

Data transformation: Transform the variables to create new, uncorrelated variables.

Ridge regression or LASSO: These regularization techniques can help mitigate multicollinearity by adding a penalty term to the regression equation.

Domain knowledge: If the correlated variables are conceptually related, you can combine them into a single composite variable.


**Multicollinearity is not always a problem. In some cases, it may not have a significant impact on the regression analysis.**

Multicollinearity may not be problematic. However, severe multicollinearity is a problem because it can increase the variance of the coefficient estimates and make the estimates very sensitive to minor changes in the model. The result is that the coefficient estimates are unstable and difficult to interpret. Multicollinearity saps the statistical power of the analysis, can cause the coefficients to switch signs, and makes it more difficult to specify the correct model.





## 2. VIF(Variation Inflation Factor),Weight of Evidence & Information Value. Why and when to use?

VIF (Variation Inflation Factor), Weight of Evidence (WoE), and Information Value (IV) are statistical measures used in different contexts for analyzing and understanding data.

### **Why we should use VIF(Variation Inflation Factor),Weight of Evidence & Information Value :-**

Variation Inflation Factor (VIF):
VIF is used in regression analysis to assess multicollinearity, which is the presence of strong correlations between predictor variables. It measures how much the variance of the estimated regression coefficient is inflated due to multicollinearity.

Weight of Evidence (WoE):
WoE is a measure used in credit scoring and predictive modeling to assess the predictive power of categorical variables. It quantifies the strength of the relationship between a categorical variable and the target variable.

Information Value (IV):
IV is closely related to WoE and is also used in credit scoring and predictive modeling. It quantifies the predictive power of an independent variable in relation to the target variable.

### **When we should use VIF(Variation Inflation Factor),Weight of Evidence & Information Value :-**

We should use VIF when we want to detect and handle multicollinearity in regression models.

Weight of Evidence (WoE) and Information Value are useful in credit scoring and predictive modeling to assess the predictive power of categorical variables, select informative variables, and understand their relationship with the target variable.





## 3. What is tradeoff between bias and variances, and the relationship between them?

The trade off between bias and variance refers to, when the model is too simple then it may be on high bias and low variance thus is error prone.
And, when the model is too complex then it may be on high variance and low bias so, on latter condition new entries will not perform well.
There, is something between those two conditions called as Tradeoff.
The main goal of the tradeoff is to find the optimal balance in the middle that minimizes both bias and variance, resulting in the best generalization performance.



## 4. How overfitting and underfitting can affect model generalization?



Overfitting and underfitting are two common problems in machine learning that impact model generalization.

Overfitting: Overfitting occurs when a model learns the training data too well and fails to generalize to new, unseen data. In this case, the model becomes too complex and captures both the underlying patterns in the data as well as the noise.

Overfitting affect model generalization because it decreased the performance of the model, poor robustness.

Underfitting: Underfitting occurs when a model is too simple or lacks the capacity to capture the underlying patterns in the data. It fails to learn the complexity of the problem and produces a model that is too generalized or oversimplified.

Underfitting affect the model generalization because of high bias, not able to find or capture the patterns and due to its poor predictive performance.
