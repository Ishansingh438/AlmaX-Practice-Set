## 2. What are the differences between Adaboost and Gradient Boosting?

The main difference between Adaboost and Gradient Boosting is :

Adaboost is a machine learning algorithm that combines multiple weak models to create a strong model. It focuses on difficult examples during training by assigning them higher weights. The weak models are trained sequentially, and their predictions are combined using a weighted voting system. Main goal of Adaboost is to improve the accuracy by leveraging the strengths of different models.

On the other hand, Gradient Boosting is a machine learning algorithm that combines weak models, like decision trees, in a step-by-step manner to create a efficient predictive model. It continuously improves the model's predictions by focusing on the mistakes made by the model in previous steps. It does so by training new models to correct those mistakes, gradually reducing errors and achieving better accuracy.

## 4. What is the basic difference between ML Interpretability and ML Explainability?

Machine Learning interpretability and explainability refer to the ability to understand and explain the behavior of a machine learning model.

And the basic difference between ML Interpretability and ML Explainability is:

Interpretability has to do with how accurately a machine learning model can associate a cause to an effect.
It involves identifying the features and relationships that the model uses to make its predictions.

Explainability, on the other hand, has to do with the ability of the parameters, often hidden in Deep Nets, to justify the results. It is how to take an ML model and explain its behavior in human terms.
